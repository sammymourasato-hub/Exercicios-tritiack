# **Exemplo - MOTOR DE INDUÇÃO TRIFÁSICO**

1. INSERÇÃO DOS DADOS DE PLACA PELO USUARIO

```python
# Potência Últil (Mecânica) ---> em Watts
PU = int(input("insira a potência últil, em KW:"))
# Rendimento ---> n = PU / P
n = (float(input("insira o fator de rendimento percentual: ")))/100
# Fator de potência ----> fp = P / S
fp = float(input("insira o fator de potência (de 0 a 1) "))
```

2. CÁLCULO DAS PONTÊNCIAS ATIVA (P), APARENTE (S) E REATIVA (Q)

```python
#n = PU / P
# potência Ativa (P)
P = PU / n
#Potência Aparente (S)
#fp = P / S
S = P / fp
# Potência Reativa (Q)
Q = (S**2 - P**2)**0.5
```

3. Apresentação dos resultados

```python
# f-string
#Limitsr od resultados a 3 casas decimais
print(f"A potência ativa será de {P:.3f} KW")
print(f"A potência aparente será de {S:.3f} KVA")
print(f"A potência reativa será de {Q:.3f} KVAr")
```
