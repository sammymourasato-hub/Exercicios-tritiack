# **Exemplo - MOTOR DE INDUÇÃO TRIFÁSICO**

1. INSERÇÃO DOS DADOS DE PLACA PELO USUARIO

```python
# Potência Últil (Mecânica) ---> em Watts
PU = int(input("insira a potência últil, em Watts:"))
# Rendimento ---> n = PU / P
n = float(input("insira o fator de rendimento (de 0 a 1): "))
# Fator de potência ----> fp = P / S
fp = float(input("insira o fator de potência (de 0 a 1) "))
```

2. CÁLCULO DAS PONTÊNCIAS ATIVA (P), APARENTE (S) E REATIVA (Q)

```python
#n = PU / P
# potência Ativa (P)
P = PU / n
#Potência Aparente (S)
#fp = P / S
S = P / fp
# Potência Reativa (Q)
Q = (S**2 - P**2)**0.5
```

3. Apresentação dos resultados

```python
# f-string
print(f"A potência ativa será de {int(P)} W")
print(f"A potência aparente será de {int(P)} VA")
print(f"A potência reativa será de {int(P)} VAr")
```
